```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('car_price_prediction_.csv')
```

```python
# Drop the columns that are not in use 
dataset = dataset.drop(['Car ID','Brand','Condition','Model'],axis=1)
```

```python
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1:]
```

```python
print(x)
```

```python
print(y)
```

```python
# Encoding the independent variable 
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[2,3])],remainder='passthrough')
# ct = ColumnTransformer(transformers = [('encoder',OneHotEncoder(),[0])],remainder='passthrough')
x =np.array(ct.fit_transform(x))
```

```python
print(x)
```

```python
dataset
```

```python
# Splitting the dataset into training and testing data 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
# Frature scaling 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train[:,-1:]=sc.fit_transform(x_train[:,-1:])
x_test[:,-1:] = sc.transform(x_test[:,-1:])
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
# Evaluate the model 
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
```

```python
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')

# Print the coefficients (Beta values) and intercept
print(f'Intercept (Beta 0): {model.intercept_}')
print(f'Coefficients (Beta values): {model.coef_}')
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mse)
print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
print(f'R-squared: {r2}')
```
